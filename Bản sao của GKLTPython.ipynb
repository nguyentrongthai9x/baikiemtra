```
# kiem tra cau 1
listA = [1, 10, 20, 5, 60, 30, 10, 9, 10]
def xoa():
  for a in listA:
    if a ==10:
      listA.remove(10)
  print(listA)
print('listA co nhung phan tu')  
print(listA)
print('sau khi xoa phan tu co gia tri 10')
xoa()        
```

```
Person = {'name': 'Quoc Nam', 'age' : 26, 'salary' : 8000, 'city': 'Tuy Hoa'}
keytoremove = ['salary', 'city']
def xoakey():
  for a in keytoremove:
    del Person[a]
  print(Person)
print('Person co nhung key sau')
print(Person)
print('Sau khi xoa theo keytoremove')
xoakey()
```
